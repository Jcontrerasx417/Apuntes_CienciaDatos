Taller ANN Multiclass 

```python
##importar librerias


import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# cargar el dataset  

url = "https://github.com/adiacla/bigdata/raw/master/riesgo.xlsx"

df = pd.read_excel(url)

df.head()
```

```python

# 2. exploracion de datos 


# Mostrar información general del dataset
# (tipos de datos, valores nulos, etc.)
df.info()

# Mostrar estadísticas básicas de las variables numéricas
df.describe()
```

```python

# 3. eliminacion de columnas irrelevantes

df = df.drop(columns=['Customer_ID', 'Name', 'SSN'])
```

```python

# 4. codificacion de variables categoricas
# Se crea un codificador de etiquetas
le = LabelEncoder()

# Se recorren todas las columnas que sean tipo texto
for col in df.select_dtypes(include='object').columns:
    
    # Se convierten a string y luego a valores numéricos
    df[col] = le.fit_transform(df[col].astype(str))
```

```python

# 5. separar features y target


# Variables predictoras (features)
X = df.drop('Credit_Score', axis=1)

# Variable objetivo (target)
y = df['Credit_Score']
```

```python

# 6. normalizacion

# Se crea el escalador
scaler = StandardScaler()

# Se normalizan las variables predictoras
X_scaled = scaler.fit_transform(X)
```

```python

# 7. reduccion de dimencionalidad (PCA)


# Se crea el modelo PCA para reducir a 10 componentes principales
pca = PCA(n_components=10)

# Se transforman los datos escalados
X_pca = pca.fit_transform(X_scaled)

# Mostrar cuánta varianza explica cada componente
print("Varianza explicada por cada componente:")
print(pca.explained_variance_ratio_)
```

```python

# 8. Division de datos 


# Se divide el dataset:
# 80% para entrenamiento
# 20% para pruebas
X_train, X_test, y_train, y_test = train_test_split(
    X_pca,
    y,
    test_size=0.2,
    random_state=42
)
```

```python

# 9. ONE HOT ENCODING del target

# Como tenemos 3 clases (0,1,2)
# se convierten a formato categórico

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python

# 10. creacion del modelo ANN


# Se crea un modelo secuencial
model = Sequential()

# Capa de entrada + primera capa oculta
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))

# Segunda capa oculta
model.add(Dense(32, activation='relu'))

# Capa de salida
# 3 neuronas porque hay 3 clases
# Softmax se usa para clasificación multiclase
model.add(Dense(3, activation='softmax'))
```

```python

# 11. compilacion del modelo


model.compile(
    optimizer='adam',                 # algoritmo de optimización
    loss='categorical_crossentropy',  # función de pérdida para multiclase
    metrics=['accuracy']              # métrica de evaluación
)
```

```python

# 12. entrenamiento del modelo


history = model.fit(
    X_train,
    y_train,
    epochs=50,          # número de iteraciones de entrenamiento
    batch_size=32,      # tamaño de lote
    validation_split=0.2
)
```

```python

# 13. evaluacion del modelo


# Evaluar el modelo con los datos de prueba
loss, accuracy = model.evaluate(X_test, y_test)

print("Accuracy del modelo:", accuracy)
```

```python

# 15. predicciones

# Se obtienen las probabilidades predichas
y_pred = model.predict(X_test)

# Se convierten a la clase con mayor probabilidad
y_pred = np.argmax(y_pred, axis=1)

# Convertir los valores reales
y_true = np.argmax(y_test, axis=1)
```

```python

# 16. matriz de confusion

print(confusion_matrix(y_true, y_pred))
```

```python

# 17. reporte de clasificacion


print(classification_report(y_true, y_pred))
```
