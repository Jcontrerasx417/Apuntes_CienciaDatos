```python


# Importar las librerías necesarias
from ultralytics import YOLO
import os

# Ruta al archivo data.yaml del dataset descargado
data_path = os.path.join(dataset.location, 'data.yaml')

# Cargar el modelo YOLOv8n
model = YOLO('yolov8n.pt')

# Entrenar el modelo
resultado = model.train(
    data=data_path,
    epochs=5,
    imgsz=640,
    project='ejecucion'
)

# El resultado se guarda en la variable 'resultado' y los archivos en la carpeta 'ejecucion'
```
