```python
import cv2
import matplotlib.pyplot as plt

# Carga la imagen (BGR)
img_bgr = cv2.imread("logo.jpg")

# Convierte a RGB para mostrar correctamente en matplotlib
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Copia y deja sólo el canal verde (R=0, B=0)
img_g_only = img_rgb.copy()
img_g_only[:, :, 0] = 0  # rojo
img_g_only[:, :, 2] = 0  # azul

# Mostrar
plt.figure(figsize=(8, 5))
plt.imshow(img_g_only)
plt.axis("off")
plt.title("capa verde")
plt.show()
```

```python
!curl -o fortnite2.jpg https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/fortnite-og-social-1920x1080-a5adda66fab9.jpg

fname2 = "fortnite2.jpg"
img2_bgr = cv2.imread(fname2)
img2_rgb = cv2.cvtColor(img2_bgr, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 6))
plt.imshow(img2_rgb)
plt.axis("off")
plt.title("Imagen descargada (RGB)")
plt.show()

img2_r_only = img2_rgb.copy()
img2_r_only[:, :, 1:] = 0

img2_g_only = img2_rgb.copy()
img2_g_only[:, :, [0, 2]] = 0

img2_b_only = img2_rgb.copy()
img2_b_only[:, :, :2] = 0

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img2_r_only); axs[0].axis("off"); axs[0].set_title("Canal rojo")
axs[1].imshow(img2_g_only); axs[1].axis("off"); axs[1].set_title("Canal verde")
axs[2].imshow(img2_b_only); axs[2].axis("off"); axs[2].set_title("Canal azul")
plt.show()
```
