# <font color="blue">Deep Learning Generativo y Discriminativo en el Contexto de GenAI y LLM </font>

En la última década, el campo de la inteligencia artificial ha sido transformado radicalmente por el auge del deep learning (aprendizaje profundo). Este paradigma ha dado lugar a dos enfoques principales: los modelos generativos y los discriminativos. Ambos tipos de modelos tienen aplicaciones fundamentales, pero desempeñan roles diferentes en la forma en que las máquinas perciben y generan conocimiento. En este ensayo se analiza la distinción entre estos dos enfoques, su integración en tecnologías emergentes como la Inteligencia Artificial Generativa (GenAI), y su expresión más avanzada en los Large Language Models (LLM).

## Modelos Discriminativos: Clasificar y Distinguir

Los modelos discriminativos se centran en aprender la frontera entre clases. Dado un conjunto de datos con entradas y etiquetas, estos modelos aprenden a predecir directamente la etiqueta de una entrada. Ejemplos típicos de modelos discriminativos en deep learning son las redes neuronales convolucionales para clasificación de imágenes, o los clasificadores de texto como BERT en tareas de sentimiento.

Desde una perspectiva matemática, los modelos discriminativos aprenden la probabilidad condicional
$P(y∣x)$, es decir, la probabilidad de una etiqueta $y$ dado un dato de entrada
$x$. Esta aproximación permite una clasificación precisa y eficiente, pero no permite generar ejemplos nuevos del dominio aprendido.

## Modelos Generativos: Crear y Simular

En contraste, los modelos generativos aprenden la distribución conjunta
$P(x,y)$, o incluso solo $P(x)$, y pueden generar datos nuevos que se asemejan al conjunto de entrenamiento. Esto los convierte en una herramienta clave para la creatividad artificial. Los modelos como GANs (Generative Adversarial Networks), VAEs (Variational Autoencoders) y modelos autoregresivos como GPT (Gerative Preentrain Transformers), son ejemplos de arquitecturas generativas.

En el contexto de texto, imágenes, audio o video, estos modelos son capaces de producir contenido original. Por ejemplo, GPT-4 puede escribir poesía, resolver problemas matemáticos o simular una conversación humana, todo a partir del conocimiento aprendido durante el entrenamiento.

#GenAI: Inteligencia Artificial Generativa

La Inteligencia Artificial Generativa (GenAI) se refiere al conjunto de tecnologías que permiten crear contenido nuevo a partir de datos existentes. Esta capacidad va más allá de la simple predicción; permite sintetizar ideas, diseñar productos, generar código, música o imágenes. GenAI se basa casi exclusivamente en modelos generativos profundos y representa uno de los avances más disruptivos en la historia reciente de la computación.

Entre sus aplicaciones más notables se encuentran:

* Creación automática de textos y documentos.

* Generación de imágenes realistas (como en DALL·E o Midjourney).

* Síntesis de voz y música.

* Diseño de medicamentos o materiales nuevos mediante simulaciones moleculares.

El impacto de GenAI no es solo técnico, sino también cultural, legal y ético, ya que plantea preguntas sobre la autoría, la verdad, la originalidad y el uso responsable de la tecnología.

# LLM: Grandes Modelos de Lenguaje como Gemmi, Deep Seek, Claude, Llama y GPT

Los Large Language Models (LLM), como GPT-4, PaLM, Claude o LLaMA, son la manifestación más avanzada de la inteligencia artificial generativa. Estos modelos, entrenados sobre vastas cantidades de texto, pueden generar respuestas coherentes, traducir lenguajes, resumir documentos y realizar tareas de razonamiento complejo.

Los LLM combinan principios tanto generativos como discriminativos. Aunque están diseñados para generar texto (una tarea generativa), utilizan mecanismos de atención y arquitecturas que permiten comprender y discriminar el contexto de manera eficiente.

Estos modelos han sido posibles gracias a avances en:

* Arquitecturas como Transformers.

* Entrenamiento masivo en datos no estructurados.

* Infraestructura computacional distribuida.

Además, la combinación de LLM con fine-tuning y prompt engineering ha permitido personalizar sus capacidades, ampliando su aplicabilidad a la industria, la educación, la medicina, el derecho y otras áreas.


Entonces, La IA generativa es una categoría que contiene una miríada de herramientas creadas para utilizar la información de los grandes modelos de lenguaje y otros tipos de modelos de IA que utilizan el aprendizaje automático para generar nuevos contenidos, mientras que un LLM es un tipo de modelo de IA que utiliza Machine Learning basado en miles de millones de parámetros para comprender y producir texto.

# <font color="blue"> PLN y Transformers  </font>

El Procesamiento del Lenguaje Natural (PLN) es una rama interdisciplinaria entre la lingüística, la inteligencia artificial y el aprendizaje automático, que tiene como propósito principal que las máquinas puedan entender, generar y razonar sobre el lenguaje humano. Lejos de limitarse a la simple identificación de palabras, el PLN busca capturar el significado contextual, las relaciones sintácticas, y los matices semánticos que hacen del lenguaje una herramienta compleja y rica.

##Principales tareas del PLN
Algunas de las tareas fundamentales del PLN incluyen:

* Clasificación de oraciones: Por ejemplo, determinar si una reseña es positiva o negativa.

* Etiquetado por palabra: Identificar si una palabra es un sustantivo, verbo, o entidad nombrada.

* Generación de texto: Completar frases, generar respuestas, o escribir descripciones.

* Respuesta a preguntas: Extraer información específica dada una pregunta y un contexto.

* Transformación de texto: Traducción automática o resumen de contenido.

El PLN también se extiende más allá del texto puro, participando en tareas como reconocimiento automático de voz y descripción de imágenes, lo que lo conecta con otros subcampos de la IA como el procesamiento de audio y visión por computador.


## Revolución de los Transformers en el PLN

A partir de 2017 (algunos textos mencionan el 2018), con la publicación del artículo Attention is All You Need, el modelo Transformer redefinió el paradigma del PLN. Su arquitectura basada en atención permitió:

* Paralelizar el entrenamiento, algo difícil con modelos RNN o LSTM.

* Capturar relaciones de largo alcance entre palabras en un texto.

* Servir de base para modelos avanzados como BERT, GPT, T5, RoBERTa, y más.

Hoy, los Transformers son el estándar de facto en casi todas las tareas de PLN.

# Pipeline().
Esta función conecta un modelo con los pasos necesarios para su preprocesamiento y posprocesamiento, permitiéndonos introducir de manera directa cualquier texto y obtener una respuesta inteligible.

```python

from transformers import pipeline

# 1. Análisis de Sentimiento
classifier = pipeline("sentiment-analysis")
print("Sentimento:", classifier("Estoy que me muero, no quiero vivir"))
```

**Se pueden pasar varias frases**

Por defecto, un pipeline selecciona un modelo particular preentrenado que ha sido ajustado para el análisis de sentimientos en Inglés. El modelo se descarga y se almacena en el caché cuando creas el objeto classifier. Si vuelves a ejecutar el comando, se usará el modelo almacenado en caché y no habrá necesidad de descargarlo de nuevo.

Hay tres pasos principales que ocurren cuando pasas un texto a un pipeline:

El texto es preprocesado en un formato que el modelo puede entender.
La entrada preprocesada se pasa al modelo.
Las predicciones del modelo son posprocesadas, de tal manera que las puedas entender.

```python
print("Sentimento en frases:", classifier(
    ["He estado esperando es curso de Alfredo Diaz toda el alma.", "¡Creo que no deberían enseñar ciencia de datos solo chatgpt o claude!"]
    ))
```

#Clasificación zero-shot
Una tarea compleja, en la que se necesita clasificar textos que no han sido etiquetados. Este es un escenario común en proyectos de la vida real porque anotar texto usualmente requiere mucho tiempo y dominio del tema.

Para este caso de uso, el pipeline zero-shot-classification es muy poderoso: permite que especifiques qué etiquetas usar para la clasificación, para que no dependas de las etiquetas del modelo preentrenado.
Ya viste cómo el modelo puede clasificar una oración como positiva o negativa usando esas dos etiquetas — pero también puede clasificar el texto usando cualquier otro conjunto de etiquetas que definas.

#Ejemplo

Este código toma una frase en lenguaje natural y la convierte en un vector numérico de 768 dimensiones usando el modelo DistilBERT. Este vector, llamado embedding, captura el significado del texto de una forma que puede ser utilizada por algoritmos de machine learning para tareas como clasificación, búsqueda semántica o análisis de sentimientos."

```python
# 2. Extracción de características (embeddings)
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")

inputs = tokenizer("Me gusta la maestría en ciencia de datos es mi preferida", return_tensors="pt")
with torch.no_grad():
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)
print("Vector shape:", embeddings.shape)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
classifier(
    "Este curso de transformers de la maestría de ciencia de datos es bueno",
    candidate_labels=["tecnología", "religión", "administración"],
)
```

# Named Entity Recognition (NER)

Este código utiliza una tubería de reconocimiento de entidades nombradas (NER) con Hugging Face para identificar entidades como organizaciones, lugares, personas, etc., dentro de un texto.



```python
# 4. Named Entity Recognition (NER)

from transformers import pipeline

ner = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english")
print("NER:", ner("unab is located in Bucaramanga city"))
```

UNAB (Organización):

'UN' (B-ORG): Inicio de la entidad Organización.

'##A' (I-ORG): Parte interna de la Organización.

'##B' (I-ORG): Parte interna de la Organización.

El modelo ha reconstruido UNAB a partir de los tokens UN, ##A, ##B. Esto es correcto.

Bucaramanga (Localización):

'Bu' (B-LOC): Inicio de la entidad Localización.

'##cara' (I-LOC): Parte interna de la Localización.

'##mang' (I-LOC): Parte interna de la Localización.

'##a' (I-LOC): Parte interna de la Localización.

El modelo ha reconstruido Bucaramanga a partir de los tokens Bu, ##cara, ##mang, ##a. Esto también es correcto.

# Pregunta y Respuesta

Crea una tubería de pregunta-respuesta (question-answering) usando un modelo preentrenado.

Le das una pregunta y un contexto donde debe buscar la respuesta.

El modelo analiza el contexto y extrae la respuesta más probable.

Imprime la respuesta, por ejemplo:

```python
# 5. Pregunta y Respuesta (Corregido con modelo abierto)
from transformers import pipeline
import numpy as np

# Usamos un modelo abierto que no requiere autenticación ni tokens
qa = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct")

# Ejemplo con un contexto más denso y una pregunta específica
pregunta_ejemplo = "¿Quién pintó la Mona Lisa?"
contexto_ejemplo = """
La Mona Lisa es una obra de arte icónica del Renacimiento, famosa por su enigmática sonrisa.
Fue pintada por el renombrado artista italiano Leonardo da Vinci en el siglo XVI.
Actualmente, se exhibe en el Museo del Louvre en París, Francia.
"""

# Estructuramos el prompt para el modelo
prompt_1 = f"Contexto: {contexto_ejemplo}\n\nPregunta: {pregunta_ejemplo}\n\nRespuesta corta:"

# Generamos la respuesta
resultado = qa(prompt_1, max_new_tokens=20, clean_up_tokenization_spaces=True)

print(f"Pregunta: {pregunta_ejemplo}")
print(f"Contexto:\n{contexto_ejemplo}")
# Extraemos solo el texto generado nuevo
print("Pregunta y respuesta:", resultado[0]['generated_text'].replace(prompt_1, "").strip())


# --- Otro ejemplo más específico ---
print("\n--- Segundo ejemplo ---")
pregunta_ejemplo_2 = "¿Cuál es la capital de Francia?"
contexto_ejemplo_2 = """
París, la capital de Francia, es conocida por su arquitectura, museos y gastronomía.
Es una de las ciudades más visitadas del mundo y alberga monumentos como la Torre Eiffel.
"""

prompt_2 = f"Contexto: {contexto_ejemplo_2}\n\nPregunta: {pregunta_ejemplo_2}\n\nRespuesta corta:"

resultado_2 = qa(prompt_2, max_new_tokens=20, clean_up_tokenization_spaces=True)

print(f"Pregunta: {pregunta_ejemplo_2}")
print(f"Contexto:\n{contexto_ejemplo_2}")
print("Pregunta y respuesta:", resultado_2[0]['generated_text'].replace(prompt_2, "").strip())
```

#Resumen automático

```python
# 6. Resumen automático
# 6. Resumen automático (Corregido para text-generation)
from transformers import pipeline

# Usamos la tarea 'text-generation' con un modelo abierto que entiende español
summarizer = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct")

texto = (
    "Los Transformers son una arquitectura de redes neuronales que revolucionaron el "
    "procesamiento de lenguaje natural (NLP). Fueron introducidos en el artículo 'Attention is All You Need' "
    "(Vaswani et al., 2017). Gracias a su mecanismo de auto-atención, pueden capturar relaciones contextuales "
    "entre palabras en una oración de manera más efectiva que modelos anteriores."
)

# Estructuramos el prompt para ordenarle al modelo que resuma el texto en español
prompt = f"Resume el siguiente texto en una sola frase corta:\n\nTexto: {texto}\n\nResumen:"

# Generamos el resumen limitando los tokens máximos
resultado = summarizer(prompt, max_new_tokens=40, clean_up_tokenization_spaces=True)

# Extraemos solo el texto generado nuevo
resumen_limpio = resultado[0]['generated_text'].replace(prompt, "").strip()
print("Resumen:", resumen_limpio)
```

# zero-shot
La clasificación zero-shot permite categorizar un texto sin necesidad de haber entrenado el modelo previamente con esas categorías.
Esto significa que puedes darle etiquetas nuevas (no vistas) y el modelo usará el significado de esas palabras para decidir cuál se ajusta mejor al texto.

Ideal cuando no tienes datos etiquetados o quieres clasificar en categorías flexibles.



```python
from transformers import pipeline

# 9. Clasificación zero-shot (sin entrenamiento previo con estas etiquetas)
clasificador = pipeline("zero-shot-classification")

texto = "La paz duradera es un beneficio para todos."
etiquetas = ["políticos", "presidentes", "ciudadanos"]

resultado = clasificador(texto, candidate_labels=etiquetas)

print("Clasificación zero-shot:", resultado)
```

#Uso de la API de Hugging Face para Inferencias sin Descargar el Modelo Localmente

El propósito de este código es utilizar un servicio de inferencia en la nube proporcionado por Hugging Face (o un proveedor compatible como Neibius) para interactuar con un modelo de lenguaje preentrenado. A través de una API REST, este código permite generar respuestas basadas en un conjunto de ingredientes dados, sin la necesidad de descargar el modelo en tu máquina local. Esto es útil cuando se desea acceder a los modelos de procesamiento de lenguaje natural (PLN) y generar respuestas de manera rápida y eficiente, utilizando recursos de computación en la nube.

En este caso, el modelo se utilizará para generar recomendaciones de recetas basadas en los ingredientes que se le proporcionen.

## Aspectos Claves

No es necesario descargar el modelo localmente:

La inferencia se realiza en la nube, lo que significa que no es necesario contar con hardware potente (como GPU o CPU de alto rendimiento) para ejecutar el modelo. Los cálculos se realizan en los servidores del proveedor de la API (en este caso, Neibius o Hugging Face).

**1. Uso de un token de autenticación:**

Para realizar solicitudes a la API, se requiere un token de API que sirve para autenticar la conexión con el servicio de inferencia. Este token es proporcionado por el servicio que utilices (como Hugging Face o Neibius).

**2. Interacción a través de un modelo preentrenado:**

El modelo utilizado en este ejemplo es el modelo Llama-3.2-3B-Instruct, que es un modelo de lenguaje grande diseñado para tareas de inferencia basadas en instrucciones. Este modelo está preentrenado en grandes volúmenes de datos y es capaz de generar respuestas detalladas basadas en las instrucciones que se le proporcionen.

**3. Generación dinámica de contenido:**

El código toma una lista de ingredientes como entrada y, a partir de esta, genera un prompt que se envía al modelo. El modelo responde con recomendaciones de platos basadas en los ingredientes dados, proporcionando incluso recetas completas.



```python

from huggingface_hub import InferenceClient

# Crea el cliente solo con tu token
client = InferenceClient(api_key="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")

ingredients_list = ['arroz', 'piña', 'leche', 'huevos']

prompt = f"Como chef profesional, ¿puedes darme dos recomendaciones de platos usando los siguientes ingredientes: {', '.join(ingredients_list)}? Proporciona las recetas completas para cada plato."

# Llamada al modelo
completion = client.chat.completions.create(
    model="meta-llama/Llama-3.2-3B-Instruct",
    messages=[
        {"role": "user", "content": prompt}
    ],
)

# Mostrar respuesta
print(completion.choices[0].message["content"])
```
