```python
from google import genai
import gradio as gr
import os

# Configurar clave de API (asegúrate de poner la tuya)
os.environ["GEMINI_API_KEY"] = "xxxxs"

# Crear cliente Gemini
client = genai.Client()

# Crear un chat persistente con el modelo
chat = client.chats.create(model="gemini-2.5-flash")

# Función de respuesta para Gradio
def responder(mensaje, historia):
    try:
        # Enviar mensaje al modelo
        respuesta = chat.send_message(mensaje)
        # Añadir al historial
        historia = historia or []
        historia.append({"role": "user", "content": mensaje})
        historia.append({"role": "assistant", "content": respuesta.text})
        return historia
    except Exception as e:
        return [{"role": "assistant", "content": f"❌ Error: {e}"}]

# Interfaz de chat con formato 'messages'
demo = gr.ChatInterface(
    fn=responder,
    title=" Chat con Gemini 2.5 Flash",
    description="Chat interactivo con el modelo Gemini más reciente (API google.genai).",
    type="messages",
    examples=[
        {"role": "user", "content": "¿Qué es la inteligencia artificial?"},
        {"role": "user", "content": "Explica la diferencia entre IA y machine learning."}
    ]
)

demo.launch(share=True)
```
