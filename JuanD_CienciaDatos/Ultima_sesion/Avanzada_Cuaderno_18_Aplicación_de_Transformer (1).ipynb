#Taller Resumen de Texto

En este ejemplo, utilizaremos la biblioteca Hugging Face Transformers para resumir un texto largo. Utilizaremos el modelo facebook/bart-large-cnn, especializado en tareas de resumen.

##Descripción del Código:

Este taller ilustra una de las múltiples aplicaciones de las redes Transformer, mostrando cómo se pueden utilizar para extraer la esencia de textos largos, lo que resulta muy útil en escenarios donde se requiere un análisis rápido de grandes volúmenes de información.

* Importación y Pipeline:
Se importa pipeline de la biblioteca transformers y se crea un pipeline para resumen utilizando el modelo facebook/bart-large-cnn.

* Entrada de Texto:
Se define un texto extenso que servirá de entrada para la tarea de resumen.

* Generación del Resumen:
El pipeline procesa el texto y devuelve un resumen, especificando parámetros como la longitud mínima y máxima del resumen.

* Salida:
Se imprime el resumen generado, demostrando la capacidad del Transformer para condensar información de manera coherente.



```python
# Crear un pipeline para resumen de texto utilizando la tarea compatible 'text-generation'
from transformers import pipeline

# Usamos un modelo abierto que no requiere tokens y entiende español perfectamente
summarizer = pipeline("text-generation", model="Qwen/Qwen2.5-1.5B-Instruct")

# Texto de entrada: un ejemplo de párrafo extenso
texto_largo = """
La Maestría en Ciencia de Datos tiene como objetivo formar profesionales que dominen los principios
fundamentales de la Ciencia de Datos, dotándoles de una sólida base científica, metodologías avanzadas y
 habilidades en el uso de modelos y algoritmos. Además, se les proporcionará un conocimiento profundo de las
  herramientas computacionales necesarias para proponer, diseñar y ejecutar proyectos basados en datos,
  que aporten un valor significativo a las empresas de todos los sectores que recolectan, procesan y toman
  decisiones informadas sobre la información.
"""

# Estructuramos el prompt para indicarle al modelo que resuma el texto
prompt = f"Resume el siguiente texto en una frase de entre 30 y 60 palabras:\n\nTexto: {texto_largo}\n\nResumen:"

# Agregamos max_length=None para evitar la advertencia de conflicto de parámetros de generación
resultado = summarizer(
    prompt,
    max_new_tokens=60,
    max_length=None,
    clean_up_tokenization_spaces=True
)

# Extraemos solo el texto nuevo generado y limpiamos los espacios
resumen = resultado[0]['generated_text'].replace(prompt, "").strip()
```

```python
print("Resumen generado:")
print(resumen)
```

# 3. Arquitectura de una Red Transformer

![imagen](https://github.com/adiacla/bigdata/blob/master/transformerarqutectura.png?raw=true)

## Explicación Teórica

La arquitectura de una red Transformer se basa en una estructura modular y repetitiva que facilita el procesamiento paralelo de secuencias. Está compuesta principalmente por dos bloques:

###Codificador (Encoder):

* Entradas y Embedding: Recibe la secuencia de entrada, la transforma en vectores de embedding y añade información posicional para conservar el orden.

* Bloques de Codificación: Consiste en una pila de capas idénticas. Cada capa tiene dos subcomponentes:

  - Atención Multi-Cabeza (Self-Attention): Permite que cada token de la secuencia se enfoque en otros tokens, capturando relaciones contextuales.

  - Red Neuronal Feed-Forward: Una red completamente conectada que se aplica de forma independiente a cada posición.

* Conexiones Residuales y Normalización: Se utilizan para facilitar el flujo del gradiente y estabilizar el entrenamiento.

###Decodificador (Decoder):

Similar en estructura al codificador, pero con dos tipos de mecanismos de atención:

* Atención enmascarada (Masked Self-Attention): Asegura que en la generación de secuencias, un token solo atienda a tokens previos.

* Atención de codificador-decodificador (Cross-Attention): Permite al decodificador enfocarse en las representaciones generadas por el codificador.

Además, cuenta con una red feed-forward y utiliza conexiones residuales y normalización en cada capa.

##Puntos Clave de la Arquitectura Transformer:

* Paralelización: Gracias al mecanismo de atención, se evita el procesamiento secuencial que caracteriza a RNNs y LSTM.
* Atención Multi-Cabeza: Permite capturar diferentes relaciones y patrones de dependencia en paralelo.
* Escalabilidad: La estructura modular facilita la ampliación del modelo (por ejemplo, aumentando el número de capas o cabezas).

Esta arquitectura ha sido clave para lograr avances significativos en tareas de procesamiento de lenguaje natural y otras áreas, debido a su eficiencia y capacidad para modelar relaciones complejas en los datos.


# Taller Implementación de un Bloque de Codificador

En este ejemplo, utilizaremos **PyTorch** para construir un bloque simple de codificador usando las clases proporcionadas en la librería, que ilustra la esencia de la arquitectura Transformer.

* Definición de Parámetros:

  Se establecen parámetros esenciales como d_model (dimensión del embedding), nhead (número de cabezas de atención), y dim_feedforward (tamaño de la red feed-forward).

* Bloque de Codificador:
  
  Se crea una capa de codificador con nn.TransformerEncoderLayer que integra la atención multi-cabeza, la red feed-forward y mecanismos de normalización y conexiones residuales.

* Pila de Capas:

  Se apila la misma capa num_layers veces para construir el codificador completo utilizando nn
  
  
**Ejemplo de Datos:**

Se genera un tensor aleatorio que simula una secuencia de entrada con la forma requerida y se procesa a través del codificador.

* Salida:

  Se imprime la forma de la salida, que debe mantener la misma forma que la entrada, demostrando la transformación aplicada por el codificador.


**torch:** El núcleo de PyTorch, que proporciona soporte para tensores (similares a los arrays de NumPy, pero con capacidades adicionales como cálculo en GPU).

**torch.nn:** Un módulo que contiene funciones y clases para construir redes neuronales, en este caso, el TransformerEncoderLayer y el TransformerEncoder.


**d_model:** Es la dimensión de los embeddings o representaciones de las palabras (en este caso, 512). Representa el tamaño de cada vector que describe un token o palabra.

**nhead:** Es el número de cabezas en la atención multi-cabeza. Esto significa cuántas "atenciones" paralelas el modelo utilizará para aprender diferentes representaciones de las entradas.

**dim_feedforward:** Es el tamaño de las capas feed-forward dentro de la capa del transformador. Aquí está definido como 2048, lo cual es típico en modelos grandes.

**dropout:** Es la tasa de dropout que se aplicará en la capa de atención y feed-forward para prevenir el sobreajuste. Un valor de 0.1 significa que el 10% de las conexiones se "descartarán" aleatoriamente durante el entrenamiento.


```python
import torch
import torch.nn as nn

# Definir un bloque de codificador utilizando TransformerEncoderLayer de PyTorch
d_model = 512   # Dimensión de embedding
nhead = 8       # Número de cabezas en la atención multi-cabeza
dim_feedforward = 2048  # Dimensión de la red feed-forward
dropout = 0.1   # Tasa de dropout
```

**d_model:** El tamaño de los embeddings (512).

**nhead:** El número de cabezas de la atención multi-cabeza (8).

**dim_feedforward:** El tamaño de la capa feed-forward (2048).

**dropout:** La tasa de dropout.

**activation="relu":** Se define la activación para las capas feed-forward como ReLU.

```python
# Crear una capa de codificador Transformer con batch_first=True
encoder_layer = nn.TransformerEncoderLayer(
    d_model=d_model,
    nhead=nhead,
    dim_feedforward=dim_feedforward,
    dropout=dropout,
    activation="relu",
    batch_first=True  # Se activa batch_first para mejorar el rendimiento
)
```

Se define un codificador transformer completo usando varias capas de TransformerEncoderLayer.

El número de capas se establece en 6, lo que significa que el codificador tendrá 6 capas de atención y procesamiento feed-forward.

Y se crea un tensor aleatorio de entrada utilizando torch.rand, que genera un tensor con valores aleatorios entre 0 y 1.

El formato del tensor es (secuencia de longitud, tamaño de batch, dimensión del embedding), es decir:

**10:** Longitud de la secuencia (pueden ser 10 tokens o palabras en una secuencia).

**32:** Tamaño del batch (número de secuencias procesadas a la vez).

**512:** Dimensión del embedding de cada token (esto coincide con d_model).


```python
# Definir el codificador completo con múltiples capas (por ejemplo, 6 capas)
num_layers = 6
transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=num_layers)

# Ejemplo de entrada: secuencia de longitud 10, batch size 32, dimensión de embedding 512
# Formato: (batch_size, seq_length, d_model)
src = torch.rand(32, 10, d_model)  # Aquí el formato cambia: batch_size, seq_length, d_model
```

Pasa el tensor de entrada src a través del transformer encoder. El codificador transformer aplicará su proceso de atención y red feed-forward a cada token en cada capa del transformador.



```python
# Pasar la entrada por el codificador
output = transformer_encoder(src)

print("Forma de la salida:", output.shape)
```

#4  Codificación: el «embedding» de entrada
![imagen](https://github.com/adiacla/bigdata/blob/master/tranformerembeding.png?raw=true)

**Explicación Teórica**

El proceso de embedding consiste en transformar cada token (por ejemplo, una palabra o símbolo) de una secuencia en un vector continuo de dimensión fija.

Esta representación densa captura características semánticas y sintácticas, permitiendo que el modelo aprenda relaciones entre palabras de manera más eficiente.

Algunos puntos clave sobre el embedding en los Transformers son:

* Conversión de Datos Discretos a Continuos:

  Cada palabra se mapea a un vector en un espacio multidimensional, lo que facilita que el modelo detecte similitudes y patrones en los datos.

* Dimensión del Embedding (d_model):

  Especifica el tamaño del vector que representa cada token. Por ejemplo, en muchos modelos Transformer se utiliza una dimensión de 512 o 768.

* Aprendizaje de Representaciones:

  La capa de embedding puede inicializarse aleatoriamente y aprenderse durante el entrenamiento, o bien utilizar embeddings pre-entrenados (por ejemplo, Word2Vec o GloVe).

* Integración en la Arquitectura Transformer:

  Antes de aplicar mecanismos de atención, la secuencia de tokens se convierte en una secuencia de vectores embedding. A esta representación se le suma la codificación posicional para mantener la información del orden de los tokens.



# Taller Uso de una Capa de Embedding con PyTorch

El siguiente ejemplo demuestra cómo utilizar una capa de embedding en PyTorch para transformar una secuencia de tokens en sus correspondientes vectores de embedding.


* Definición de Parámetros:
  
  Se define el tamaño del vocabulario (vocab_size) y la dimensión del embedding (d_model), que determinan la cantidad de tokens únicos y la longitud del vector de cada token, respectivamente.

* Creación de la Capa de Embedding:
  
  Utilizamos nn.Embedding de PyTorch para transformar índices de tokens en vectores continuos.

* Transformación de una Secuencia de Tokens:

  Se simula una secuencia de tokens representada como una lista de índices. Al pasarla por la capa de embedding, cada índice se convierte en un vector de dimensión d_model.

**Salida del Ejemplo:**

Se imprime la forma del tensor resultante, que debería tener la forma (longitud_secuencia, d_model), demostrando la conversión de la secuencia de tokens a su representación embebida.


```python
# Asegúrate de tener instalada la librería PyTorch:
# pip install torch

import torch
import torch.nn as nn

# Parámetros del embedding:
vocab_size = 10000     # Tamaño del vocabulario (número total de tokens únicos)
d_model = 512          # Dimensión del embedding

# Crear la capa de embedding
embedding_layer = nn.Embedding(num_embeddings=vocab_size, embedding_dim=d_model)

# Ejemplo de secuencia de tokens (índices de palabras)
# Supongamos una secuencia de 10 tokens
tokens = torch.tensor([1, 345, 678, 23, 890, 12, 0, 456, 789, 2], dtype=torch.long)

# Aplicar la capa de embedding a la secuencia
embedded_tokens = embedding_layer(tokens)

print("Forma de la salida:", embedded_tokens.shape)
```

# 5 Codificación: el codificador posicional
![imagen](https://github.com/adiacla/bigdata/blob/master/transformerposicional.png?raw=true)

**Explicación Teórica**

En la arquitectura Transformer, a diferencia de las redes recurrentes, no se procesa la información de manera secuencial, lo que significa que no existe una noción inherente del orden de los tokens. Para incorporar esta información de orden, se utiliza la codificación posicional. Esta técnica añade una representación única de posición a cada token, permitiendo que el modelo distinga entre la posición relativa y absoluta de cada elemento en la secuencia.

**Características principales:**

* Objetivo:

  Proveer a cada token información acerca de su posición en la secuencia, complementando el embedding que sólo representa características semánticas.

* Funcionamiento:

  Se utiliza una combinación de funciones sinusoidales de diferentes frecuencias. Las fórmulas son:

$$PE_{(𝑝𝑜𝑠,2𝑖)}=sin(\frac{pos}{10000^{\frac{2i}{d\_model}}})$$

$$PE_{(𝑝𝑜𝑠,2𝑖)}=cos(\frac{pos}{10000^{\frac{2i}{d\_model}}})$$

Donde:

  * $pos$ es la posición del token en la secuencia.

  * $i$ es el índice de la dimensión.

  * $d\_model$ es la dimensión del embedding.

Ventajas:

*No requiere aprendizaje:* Las funciones sinusoidales están definidas matemáticamente.

*Generalización:* Permite extrapolar a secuencias de longitudes diferentes a las vistas durante el entrenamiento.

#Taller Generación de Positional Encoding

El siguiente ejemplo utiliza PyTorch para calcular la codificación posicional para una secuencia dada:


**Función positional_encoding:**

Calcula la codificación posicional para cada token de la secuencia. Utiliza funciones sin y cos para generar representaciones únicas basadas en la posición y la dimensión.

**Parámetros:**

* seq_len: La longitud de la secuencia (número de tokens).

* d_model: La dimensión del vector de embedding.

**Cálculo de div_term:**

Este tensor permite asignar diferentes frecuencias a cada dimensión, garantizando que cada posición tenga una representación única.

**Salida:**
Se imprime la forma del tensor pe, que será de la forma (seq_len, d_model), confirmando que cada token tiene su correspondiente vector de codificación posicional.



```python
# Asegúrate de tener instalada la librería PyTorch:
# pip install torch

import torch
import math

def positional_encoding(seq_len, d_model):
    """
    Calcula la codificación posicional para una secuencia de longitud seq_len y dimensión d_model.
    """
    # Inicializar el tensor de codificación posicional con ceros
    pe = torch.zeros(seq_len, d_model)
    # Crear un tensor que representa las posiciones de la secuencia
    position = torch.arange(0, seq_len, dtype=torch.float).unsqueeze(1)
    # Calcular los términos de escala para las funciones sinusoidales
    div_term = torch.exp(torch.arange(0, d_model, 2).float() * -(math.log(10000.0) / d_model))

    # Aplicar sin para índices pares y cos para índices impares
    pe[:, 0::2] = torch.sin(position * div_term)  # para dimensiones pares
    pe[:, 1::2] = torch.cos(position * div_term)  # para dimensiones impares
    return pe

# Parámetros
seq_len = 50   # Longitud de la secuencia
d_model = 512  # Dimensión del embedding

# Calcular la codificación posicional
pe = positional_encoding(seq_len, d_model)
print("Forma de la codificación posicional:", pe.shape)
```

#6. Codificación: el bloque atencional

![imagen](https://github.com/adiacla/bigdata/blob/master/tranformermatrizselfatencion.png?raw=true)

##Explicación Teórica

El bloque atencional es uno de los componentes centrales en las redes Transformer. Su función es permitir que cada token en una secuencia “atienda” (o se enfoque) en otros tokens, de modo que se puedan capturar relaciones contextuales a lo largo de toda la secuencia.

##Puntos clave del bloque atencional:

**Mecanismo de Self-Attention:**

Cada token actúa simultáneamente como consulta (query), clave (key) y valor (value). El mecanismo calcula la atención entre todos los pares de tokens, permitiendo que el modelo determine la relevancia de cada token con respecto a los demás.


**Scaled Dot-Product Attention:**

La atención se calcula mediante el producto escalar entre las consultas y las claves. Para estabilizar los gradientes, este producto se escala dividiéndolo por la raíz cuadrada de la dimensión de las claves $(\sqrt{𝑑_k})$
La fórmula básica es:

$$
Attention(Q,K,V)=softmax(\frac{QK^T}{\sqrt{𝑑_k}} )V
$$
donde:

* Q es la matriz de consultas.

* K es la matriz de claves.

* V es la matriz de valores.

**Multi-Head Attention (múltiples cabezas):**

En la práctica, en lugar de realizar una única atención, se dividen los vectores en varias “cabezas”. Cada cabeza aprende una representación distinta del espacio de atención, lo que permite capturar diferentes tipos de relaciones y patrones en los datos.

*Ventajas:*

* Permite capturar relaciones a largo plazo sin depender del orden secuencial.

* Ofrece un mecanismo paralelo y escalable para procesar secuencias.

* Mejora la capacidad del modelo para representar información contextual.

# Taller Implementación del Scaled Dot-Product Attention

En el siguiente ejemplo, implementamos el mecanismo de Scaled Dot-Product Attention utilizando PyTorch. Este ejemplo ilustra cómo se calculan los scores de atención y cómo se generan las salidas atencionales a partir de las matrices de consulta, clave y valor.

**Cálculo de Scores:**

Se multiplica la matriz de consultas 𝑄 por la transpuesta de la matriz de claves 𝐾 y se escala el resultado dividiéndolo por la raíz cuadrada de $𝑑_𝑘$.

**Normalización con Softmax:**

Se aplica la función softmax a los scores para obtener probabilidades que indican la importancia relativa de cada token respecto a los demás.

**Cálculo de la Salida:**

Los pesos de atención se utilizan para ponderar la matriz de valores 𝑉, generando la salida final del bloque atencional.

Ejemplo Práctico:
Se crean tensores aleatorios para 𝑄, 𝐾 y 𝑉 y se aplica la función de atención para demostrar su funcionamiento.



```python
# Asegúrate de tener instalada la librería PyTorch:
# pip install torch

import torch
import torch.nn.functional as F
import math

def scaled_dot_product_attention(Q, K, V):
    """
    Calcula la atención mediante el mecanismo de scaled dot-product.

    Parámetros:
    - Q: Matriz de consultas de forma (batch_size, seq_length, d_k)
    - K: Matriz de claves de forma (batch_size, seq_length, d_k)
    - V: Matriz de valores de forma (batch_size, seq_length, d_v)

    Retorna:
    - output: Salida de la atención
    - attn_weights: Pesos de atención (softmax aplicado a los scores)
    """
    # Obtener la dimensión de las claves (d_k)
    d_k = Q.size(-1)

    # Calcular los scores de atención: producto punto entre Q y K^T, escalado por sqrt(d_k)
    scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(d_k)

    # Aplicar softmax para obtener los pesos de atención
    attn_weights = F.softmax(scores, dim=-1)

    # Multiplicar los pesos de atención por los valores V
    output = torch.matmul(attn_weights, V)

    return output, attn_weights

# Ejemplo de uso:
batch_size = 1
seq_length = 4
d_k = 8   # Dimensión de consultas y claves
d_v = 8   # Dimensión de los valores

# Generar tensores aleatorios para Q, K y V
Q = torch.rand(batch_size, seq_length, d_k)
K = torch.rand(batch_size, seq_length, d_k)
V = torch.rand(batch_size, seq_length, d_v)

# Calcular la atención
output, attn_weights = scaled_dot_product_attention(Q, K, V)

print("Salida de la atención (output):", output)

```

#7 Codificación: etapa de salida

## Explicación Teórica

Una vez que la entrada ha sido procesada a través de múltiples bloques de codificación en la red Transformer, se obtiene una salida final que consiste en una secuencia de representaciones contextuales para cada token. Esta etapa de salida es crucial porque:

**Resumen de la Información Contextual:**

Cada token en la salida contiene información enriquecida y contextualizada, la cual ha sido refinada a través de mecanismos de atención, conexiones residuales y normalización en cada capa del encoder.

**Preparación para la Tarea Específica:**

En modelos secuencia a secuencia (p.ej., traducción):

La salida del encoder se transfiere al decodificador a través del mecanismo de atención cruzada para generar la secuencia de salida.

En tareas de clasificación:

Se puede extraer un token especial (como [CLS] en BERT) o aplicar técnicas de pooling sobre la secuencia para obtener una representación global, la cual se pasa a una capa lineal que produce las predicciones finales.

Transformación Final (Head):

Antes de entregar la salida a la siguiente etapa (ya sea el decodificador o una capa de salida para clasificación), es habitual aplicar una capa lineal, dropout u otras transformaciones que ajusten las dimensiones y ayuden a regularizar el modelo.

Esta etapa de salida es esencial para adaptar la rica representación aprendida por el encoder al objetivo específico del modelo, ya sea generar texto, clasificar o cualquier otra tarea.

#Taller 8 Decodificación: principio de funcionamiento

![imagen](https://github.com/adiacla/bigdata/blob/master/transformerdecoder.JPG?raw=true)

##Explicación Teórica

La etapa de decodificación en una red Transformer es responsable de generar la salida (por ejemplo, una secuencia de palabras) a partir de las representaciones contextuales producidas por el encoder y la información de la secuencia generada hasta el momento. Este proceso se rige por dos mecanismos clave:

**Atención enmascarada (Masked Self-Attention):**


  Durante la generación de la secuencia, el decodificador debe evitar "mirar hacia adelante". Para lograr esto, se utiliza una máscara que impide que cada token atienda a tokens futuros, asegurando que la predicción de cada posición se base únicamente en los tokens anteriores y en la información proveniente del encoder.

**Atención de codificador-decodificador (Cross-Attention):**

  Además de procesar la secuencia de salida generada, el decodificador incorpora la información contextual del encoder mediante un mecanismo de atención que permite enfocarse en partes relevantes de la entrada original. Esto conecta la representación interna de la entrada con la generación de la salida.

##Flujo General de la Decodificación:

**Entrada del Decodificador:**

Se toma la secuencia de tokens generados hasta el momento (durante el entrenamiento se puede utilizar teacher forcing).

**Masked Self-Attention:**

Se aplica un mecanismo de autoatención con una máscara que impide la influencia de futuros tokens.

**Cross-Attention:**

Se incorpora la salida del encoder mediante atención, permitiendo que el decodificador se "fije" en partes específicas de la entrada.

**Red Feed-Forward y Normalización:**

Cada capa del decodificador incluye una red neuronal completamente conectada y mecanismos de normalización y conexiones residuales, similares a los del encoder.

Este diseño permite que el decodificador genere salidas coherentes y contextualizadas, manteniendo la integridad de la secuencia generada y aprovechando la información global proporcionada por el encoder.

#Implementación de un Decodificador Simple

En este ejemplo, utilizaremos PyTorch para ilustrar el funcionamiento de un decodificador Transformer. Se utilizará la clase nn.TransformerDecoder y nn.TransformerDecoderLayer para construir un decodificador con atención enmascarada y atención cruzada.


Generación de la Máscara:
La función generate_square_subsequent_mask crea una máscara triangular superior que impide que cada posición del decodificador vea posiciones futuras, asegurando la causalidad en la generación de la secuencia.

Definición del Decodificador:
Se define una capa de decodificador (TransformerDecoderLayer) y se apilan varias de ellas usando nn.TransformerDecoder.

Simulación de Entradas:

tgt: Simula la secuencia de tokens generados (entrada del decodificador).
memory: Simula la salida del encoder que será utilizada en la atención cruzada.
Aplicación de la Máscara y Cálculo de la Salida:
Se aplica la máscara a la secuencia del decodificador y se genera la salida final, que mantiene la forma (target_seq_len, batch_size, d_model).



```python
# Asegúrate de tener instalada la librería PyTorch:
# pip install torch

import torch
import torch.nn as nn

def generate_square_subsequent_mask(sz):
    """
    Genera una máscara triangular superior para evitar que cada posición vea posiciones futuras.
    """
    mask = torch.triu(torch.ones(sz, sz) * float('-inf'), diagonal=1)
    return mask

# Parámetros del decodificador:
d_model = 512             # Dimensión de embedding
nhead = 8                 # Número de cabezas en la atención multi-cabeza
dim_feedforward = 2048    # Dimensión de la red feed-forward
dropout = 0.1             # Tasa de dropout
num_layers = 6            # Número de capas en el decodificador

# Crear una capa de decodificador Transformer
decoder_layer = nn.TransformerDecoderLayer(d_model, nhead, dim_feedforward, dropout)
transformer_decoder = nn.TransformerDecoder(decoder_layer, num_layers=num_layers)

# Ejemplo de entradas:
target_seq_len = 10       # Longitud de la secuencia objetivo (decodificador)
src_seq_len = 15          # Longitud de la secuencia del encoder
batch_size = 32

# Simulación de la secuencia de salida (tokens generados) para el decodificador:
# Formato: (target_seq_len, batch_size, d_model)
tgt = torch.rand(target_seq_len, batch_size, d_model)

# Simulación de la memoria del encoder:
# Formato: (src_seq_len, batch_size, d_model)
memory = torch.rand(src_seq_len, batch_size, d_model)

# Generar la máscara para la secuencia del decodificador (evita mirar hacia adelante)
tgt_mask = generate_square_subsequent_mask(target_seq_len)

# Pasar las entradas por el decodificador
output = transformer_decoder(tgt, memory, tgt_mask=tgt_mask)

print("Forma de la salida del decodificador:", output.shape)
# La salida tendrá la forma: (target_seq_len, batch_size, d_model)
```
