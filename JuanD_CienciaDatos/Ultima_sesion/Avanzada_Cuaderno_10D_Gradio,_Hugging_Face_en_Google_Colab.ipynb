## 1. Gradio

**Gradio** es una biblioteca de Python que permite crear aplicaciones web interactivas para tus modelos de Machine Learning de forma sencilla. Solo necesitas definir los componentes de entrada y salida que tu modelo requiere, y Gradio se encarga del resto.

**Por favor revisar el material de estudio sobre Gradio**

"Componentes de entrada y salida", en Gradio incluye una gran variedad de componentes predefinidos que se ajustan a diferentes tipos de modelos.

Por ejemplo:

* **Clasificador de imágenes:** la entrada es una imagen (Image) y la salida es una etiqueta (Label).

* **Modelo de reconocimiento de voz:** la entrada puede ser un micrófono (Microphone, que permite grabar desde el navegador) o un archivo de audio (Audio, que permite cargar archivos), y la salida suele ser texto (Text).

* **Modelo de respuesta a preguntas**: se esperan dos entradas de texto ([Text, Text]): una para el párrafo y otra para la pregunta. La salida es un solo texto que corresponde a la respuesta generada.

Además de los componentes de entrada y salida, Gradio necesita un tercer elemento: la función de predicción. Esta puede ser cualquier función de Python que reciba los valores correspondientes a los inputs definidos y devuelva la salida esperada.

# Creación de un demo simple con Gradio

Este ejemplo muestra cómo crear una interfaz web básica usando la biblioteca **Gradio**.

La función `gr.Interface()` permite construir una aplicación interactiva especificando tres elementos principales:

- `fn`: la función de Python que se desea ejecutar (en este caso, `greet`).
- `inputs`: el tipo de entrada que el usuario proporcionará (por ejemplo, un cuadro de texto).
- `outputs`: el tipo de salida que mostrará el resultado (también un cuadro de texto).


```python

import gradio as gr

gr.Interface(fn=greet, inputs="text", outputs="text").launch();
```

# Generador de Tonos Musicales con Gradio

Este ejemplo utiliza **Gradio** para crear una interfaz web interactiva que permite generar tonos musicales a partir de tres parámetros:

- **Nota**: Selección de la nota musical (C, C#, D, etc.).
- **Octava**: Número de octava (entre 4 y 6).
- **Duración**: Tiempo en segundos que debe durar el tono generado.

La función `generate_tone` genera una onda sinusoidal correspondiente a la frecuencia de la nota seleccionada y la convierte en un archivo de audio reproducible en el navegador.

Este ejemplo es ideal para explorar conceptos como frecuencias musicales, ondas senoidales, y generación de audio con Python.

>  Asegúrate de tener instalada la última versión de Gradio. Puedes actualizarla con:  

> `!pip install -U gradio`


```python
import numpy as np
import gradio as gr

# Diccionario de notas a índice (A = 9 en sistema de 12 semitonos)
note_to_index = {
    "C": 0, "C#": 1, "D": 2, "D#": 3, "E": 4,
    "F": 5, "F#": 6, "G": 7, "G#": 8, "A": 9, "A#": 10, "B": 11
}

def generate_tone(note, octave, duration):
    sampling_rate = 48000
    note_index = note_to_index[note]
    tones_from_a4 = 12 * (octave - 4) + (note_index - 9)
    frequency = 440 * 2 ** (tones_from_a4 / 12)

    duration = float(duration)
    t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
    audio = 0.5 * np.sin(2 * np.pi * frequency * t).astype(np.float32)

    return (sampling_rate, audio)

with gr.Blocks() as demo:
    gr.Markdown("## ¡Genera un tono musical!")

    note = gr.Dropdown(["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"], label="Nota", value="A")
    octave = gr.Slider(minimum=4, maximum=6, step=1, label="Octava", value=4)
    duration = gr.Number(label="Duración en segundos", value=1)

    output = gr.Audio(label="Tono generado", type="numpy")

    btn = gr.Button("Generar tono")
    btn.click(fn=generate_tone, inputs=[note, octave, duration], outputs=output)

demo.launch()
```

```python
import numpy as np
import gradio as gr

# Diccionario de notas a índice (C = 0 en sistema de 12 semitonos)
note_to_index = {
    "C": 0, "C#": 1, "D": 2, "D#": 3, "E": 4,
    "F": 5, "F#": 6, "G": 7, "G#": 8, "A": 9, "A#": 10, "B": 11
}
# Diccionario inverso para convertir índice a nota
index_to_note = {v: k for k, v in note_to_index.items()}

def generate_tone(note, octave, duration):
    sampling_rate = 48000
    note_index = note_to_index[note]
    # Cálculo de tonos desde A4 (440 Hz) para obtener la frecuencia
    tones_from_a4 = 12 * (octave - 4) + (note_index - note_to_index["A"])
    frequency = 440 * 2 ** (tones_from_a4 / 12)

    duration = float(duration)
    t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
    audio = 0.5 * np.sin(2 * np.pi * frequency * t).astype(np.float32)

    return (sampling_rate, audio) # Corrected to return tuple (sampling_rate, audio)

def generate_happy_birthday(start_note_name):
    sampling_rate = 48000
    full_audio = np.array([], dtype=np.float32)

    # Definición de la melodía base de 'Happy Birthday' comenzando en G4
    # Formato: (note_name, octave, duration_multiplier)
    # Duración base de un cuarto de nota es 0.5 segundos. Las demás son múltiplos.
    base_melody = [
        ("G", 4, 0.5), ("G", 4, 0.25), ("A", 4, 0.5), ("G", 4, 0.5), ("C", 5, 0.5), ("B", 4, 1.0),
        ("G", 4, 0.5), ("G", 4, 0.25), ("A", 4, 0.5), ("G", 4, 0.5), ("D", 5, 0.5), ("C", 5, 1.0),
        ("G", 4, 0.5), ("G", 4, 0.25), ("G", 5, 0.5), ("E", 5, 0.5), ("C", 5, 0.5), ("B", 4, 0.5), ("A", 4, 1.0),
        ("F", 5, 0.5), ("F", 5, 0.25), ("E", 5, 0.5), ("C", 5, 0.5), ("D", 5, 0.5), ("C", 5, 1.5)
    ]

    # Nota y octava de inicio de la melodía base
    base_start_note_name = "G"
    base_start_octave = 4
    base_start_abs_semitone = note_to_index[base_start_note_name] + base_start_octave * 12

    # Nota de inicio elegida por el usuario (asumimos octava 4 para la nota base del usuario)
    user_start_abs_semitone = note_to_index[start_note_name] + 4 * 12

    # Cálculo del desplazamiento en semitonos para la transposición
    semitone_shift = user_start_abs_semitone - base_start_abs_semitone

    # Duración base para una negra (quarter note)
    quarter_note_duration = 0.5

    for note_name, octave, duration_multiplier in base_melody:
        # Calcular el valor absoluto del semitono de la nota original (desde C0)
        original_abs_semitone = note_to_index[note_name] + octave * 12

        # Aplicar el desplazamiento de semitonos
        shifted_abs_semitone = original_abs_semitone + semitone_shift

        # Convertir de nuevo a la nueva nota y octava
        new_octave = shifted_abs_semitone // 12
        new_note_index = shifted_abs_semitone % 12
        new_note_name = index_to_note[new_note_index]

        # Generar el tono para la nota transpuesta
        # generate_tone now returns (sampling_rate, audio), so we extract audio[1]
        _, note_audio = generate_tone(new_note_name, new_octave, duration_multiplier * quarter_note_duration)
        full_audio = np.concatenate((full_audio, note_audio))

        # Añadir una pequeña pausa entre notas para una mejor articulación
        pause_duration = 0.05 * duration_multiplier
        pause_samples = int(sampling_rate * pause_duration)
        full_audio = np.concatenate((full_audio, np.zeros(pause_samples, dtype=np.float32)))

    return (sampling_rate, full_audio)

# Interfaz de Gradio
with gr.Blocks() as demo:
    gr.Markdown("## ¡Genera un tono musical o la canción de cumpleaños!")

    with gr.Tab("Generar Tono Individual"):
        single_note = gr.Dropdown(list(note_to_index.keys()), label="Nota", value="A")
        single_octave = gr.Slider(minimum=4, maximum=6, step=1, label="Octava", value=4)
        single_duration = gr.Number(label="Duración en segundos", value=1)
        single_output = gr.Audio(label="Tono generado", type="numpy")
        single_btn = gr.Button("Generar Tono")
        single_btn.click(fn=generate_tone, inputs=[single_note, single_octave, single_duration], outputs=single_output)

    with gr.Tab("Generar Canción de Cumpleaños"):
        song_start_note = gr.Dropdown(list(note_to_index.keys()), label="Nota de inicio de la canción", value="G")
        song_output = gr.Audio(label="Canción de Cumpleaños", type="numpy")
        song_btn = gr.Button("Generar Canción")
        song_btn.click(fn=generate_happy_birthday, inputs=[song_start_note], outputs=song_output)

demo.launch()
```

# Generación de texto en español con Gradio y Hugging Face

Este ejemplo muestra cómo generar texto en español usando un modelo de lenguaje preentrenado alojado en Hugging Face (`datificate/gpt2-small-spanish`).

- Se utiliza `gr.Interface.load()` para cargar la demo ya configurada del modelo.
- Se definen ejemplos en español para que el usuario pueda probar frases iniciales.
- La interfaz se lanza directamente en el navegador con `.launch()`.

Si encuentras un error de autorización, puede que necesites autenticarte con tu token de Hugging Face.


```python
from transformers import pipeline
import gradio as gr

# Crear pipeline de generación de texto en español
generator = pipeline("text-generation", model="datificate/gpt2-small-spanish")
#Helsinki-NLP/opus-mt-en-es o PlanTL-GOB-ES/gpt2-spanish

def generar_texto(prompt):
    resultado = generator(prompt, max_length=100, num_return_sequences=1)
    return resultado[0]["generated_text"]

# Interfaz Gradio
demo = gr.Interface(
    fn=generar_texto,
    inputs=gr.Textbox(lines=2, placeholder="Escribe texto en español..."),
    outputs="text",
    title="Generador de texto en español con GPT-2"
)

demo.launch()
```
